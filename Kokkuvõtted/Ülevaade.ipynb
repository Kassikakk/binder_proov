## Ect. haloalkaliphila LH2 isol, p = atm ja ~1.3 kbar.  
Siin on selgesti näha erinevus Car piirkonna neeldumise ja ergastuse vahel.  
Neeldumise probleem on läbivalt ebaselge taustajoon. Rakus see üldiselt servades kipub tõusma, eriti sinises e. Car pool. Samuti on mingeid ebaselgeid struktuure, nt. laiem struktuur 900 nm juures, kitsad jooned 704 ja 712 nm juures ja üldse on 600 ja 800 joonte vahel mingeid struktuure. Kitsad jooned võivad ka rubiinist olla, ehkki ei tundu, et need otseselt optilisele teele jääks. On muidugi ka eraldi mõõdetud taust rakus olemas, aga sellegi lahutamine pole kaugeltki triviaalne tegevus.  
Praegu pole näha 870 nm punasemasse minevat osa, kuigi sealt on ka mõõdetud (820 - 940 nm), aga algusosas ilmnevate taustamõjude tõttu tuleks seda veel analüüsida ja midagi maha lahutada.  
Neeldumised on hetkel tuimalt 1-T esitusse ümber arvutatud (seda tähistab "aps" (= absorptance) legendis), mis muidugi ei pruugi päris õige olla, aga siin pole kuigi lihtne leida mingit vihjet efektiivse paksuse kohta. See esindab maksimaalset vähenemist, tõde on ilmselt kuskil selle ja algse neeldumise vahepeal.  
Iseenesest võiks Car piirkonnas toimuvat ju vaadata küll nii, et vähenemist on rõhu all vähem toimunud, sest see osa on kõrgemal.


```python
import numpy as np
import pandas as pd
from math import log
import plotly.graph_objects as go

excit_p= pd.read_csv("../210406 Ect + rõhk/210406_excit_p.csv", sep='\t')
excit_0= pd.read_csv("../210330 Ect mõõtmine/210330_excit.csv", sep='\t')

abs_p= pd.read_csv("../210406 Ect + rõhk/210406_abs_p.csv", sep='\t')
abs_0= pd.read_csv("../210406 Ect + rõhk/210406_abs.csv", sep='\t')
aps_0= (1- 10**(-abs_0['abs']))/log(10) 
aps_p= (1- 10**(-abs_p['abs']))/log(10) 


fig1 = go.Figure() 
fig1.add_trace(go.Scatter(x=excit_0['wl'], y=excit_0['Excit']*1.69, error_y = dict(type = 'data', array = excit_0['Excit-Err']*1.69, visible = True), mode = 'lines + markers', name = "ex p=atm"))# NB: skaleeritud
fig1.add_trace(go.Scatter(x =excit_p['wl'], y=excit_p['Excit']/1.12, error_y = dict(type = 'data', array = excit_p['Excit-Err'], visible = True), mode = 'lines + markers', name = "ex p = 1.3 kbar")) # NB: skaleeritud
#fig1.add_trace(go.Scatter(x =abs_0['wl'], y=abs_0['abs'], mode = 'lines', name = "abs p = atm"))
fig1.add_trace(go.Scatter(x =abs_0['wl'], y=aps_0, mode = 'lines', name = "aps p = atm"))
fig1.add_trace(go.Scatter(x =abs_0['wl'], y=aps_p, mode = 'lines', name = "aps p = 1.3 kbar"))
#fig1.add_trace(go.Scatter(x =abs_p['wl'], y=abs_p['abs'], mode = 'lines', name = "abs p = 1.3 kbar"))
fig1.update_layout(height = 700, title = "Ect. hal neeldumised ja ergastused",template = "plotly_dark")
fig1
```

## Rb. sphaeroides LH2 neuro (isol) p = atm ja ~2.5 kbar
Rõhk on siin suurem ja võimalik, et neeldumise ja ergastuse erinevus osalt tuleneb ka aeglastest muutustest objektis, sest neeldumine on võetud alguses. Ilmselt oleks parem jääda veidi väiksemate rõhkude juurde.


```python

excit_p= pd.read_csv("../210414 neuroLH2 + rõhk jätk/210414_excit_p.csv", sep='\t')
excit_0= pd.read_csv("../210413 neuroLH2 + rõhk/210413_excit.csv", sep='\t')

abs_p= pd.read_csv("../210414 neuroLH2 + rõhk jätk/210414_abs_p.csv", sep='\t')
abs_0= pd.read_csv("../210413 neuroLH2 + rõhk/210413_abs.csv", sep='\t')
aps_0= (1- 10**(-abs_0['abs']))/log(10) 
aps_p= (1- 10**(-abs_p['abs']))/log(10) 
#abs_0= pd.read_csv("../210330 Ect mõõtmine/210330_excit.csv", sep='\t')

fig2 = go.Figure(data=go.Scatter(x=excit_0['wl'], y=excit_0['Excit']/1.15,error_y = dict(type = 'data', array = excit_0['Excit-Err'], visible = True), mode = 'lines + markers', name = "ex p=atm")) 
fig2.add_trace(go.Scatter(x =excit_p['wl'], y=excit_p['Excit']/1.53, error_y = dict(type = 'data', array = excit_p['Excit-Err'], visible = True), mode = 'lines + markers', name = "ex p = 2.5 kbar"))
#fig2.add_trace(go.Scatter(x =abs_0['wl'], y=abs_0['abs'], mode = 'lines', name = "abs p = atm"))
#fig2.add_trace(go.Scatter(x =abs_p['wl'], y=abs_p['abs'], mode = 'lines', name = "abs p = 2.5 kbar"))
fig2.add_trace(go.Scatter(x =abs_0['wl'], y=aps_0, mode = 'lines', name = "aps p = atm"))
fig2.add_trace(go.Scatter(x =abs_p['wl'], y=aps_p, mode = 'lines', name = "aps p = 2.5 kbar"))
fig2.update_layout(height = 700, title = "R. sp. neuroLH2 neeldumised ja ergastused",template = "plotly_dark")
fig2
```

## Acidophila LH2 isol p = atm ja ~1 kbar
Siin on ergastuses registreeritud ainult Chirascani osa, 400 - 720 nm. Proovisin sellist võtet, et jagada lihtsalt ilma lisanormeeringuta läbi vastavad ergastusspektrid eri rõhkudel.
Siin on ergastusvõimsus mõlemal juhul üpris sama, seega peaks see andma pildi vähenemise kohta. Veidi segab küll spektraalne rõhunihe, aga tendents peaks näha olema. Siin kadus kiirgus rõhu all üsna kiiresti, nii et tulemus on kohati mürane, aga tundub siiski,et karotenoidide osas on suhteline vähenemine pigem suurem, vastupidi meie eeldustele.

```python
from plotly.subplots import make_subplots
fig3 = make_subplots(specs=[[{"secondary_y": True}]])

excit_p= pd.read_csv("../210428 Acidophila/210428_excit_1.csv", sep='\t')
excit_0= pd.read_csv("../210428 Acidophila/210428_excit_0.csv", sep='\t')

rat = excit_0['Excit']/excit_p['Excit']
rat_err = rat*(excit_0['Excit-Err']/excit_0['Excit'] + excit_p['Excit-Err']/excit_p['Excit'])


abs_p= pd.read_csv("../210428 Acidophila/210428_abs_p.csv", sep='\t')
abs_0= pd.read_csv("../210428 Acidophila/210428_abs_0.csv", sep='\t')
aps_0= (1- 10**(-abs_0['abs']))/log(10) 
aps_p= (1- 10**(-abs_p['abs']))/log(10) 

fig3.add_trace(go.Scatter(x=excit_0['wl'], y=excit_0['Excit'],  error_y = dict(type = 'data', array = excit_0['Excit-Err'], visible = True),mode = 'lines + markers', name = "ex p=atm"))
fig3.add_trace(go.Scatter(x =excit_p['wl'], y=excit_p['Excit'], error_y = dict(type = 'data', array = excit_p['Excit-Err'], visible = True), mode = 'lines + markers', name = "ex p = 1 kbar"))

#fig3.add_trace(go.Scatter(x =abs_0['wl'], y=abs_0['abs'], mode = 'lines', name = "abs p = atm"),secondary_y = True)
#fig3.add_trace(go.Scatter(x =abs_p['wl'], y=abs_p['abs'], mode = 'lines', name = "abs p = 2.5 kbar"),secondary_y = True)
fig3.add_trace(go.Scatter(x =abs_0['wl'], y=aps_0, mode = 'lines', name = "aps p = atm"),secondary_y = True)
fig3.add_trace(go.Scatter(x =abs_p['wl'], y=aps_p, mode = 'lines', name = "aps p = 1 kbar"),secondary_y = True)
fig3.layout.yaxis.range = [0,8000]
fig3.layout.yaxis2.range = [0,0.22]
fig3.update_layout(height = 700, title = "Acidophila neeldumised ja ergastused",template = "plotly_dark")
fig3

```
